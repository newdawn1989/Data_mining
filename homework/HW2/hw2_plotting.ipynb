```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
import pylab

HD_data = pd.read_csv("/home/nathan/Documents/CSCI/Data_mining/homework/HW2/HDHistoricalQuotes.csv")

HD_data = HD_data.iloc[::-1]

HD_data.set_index('date', inplace=True)

HD_high=HD_data['high']
HD_low=HD_data['low']
HD_low.plot(figsize=(16, 12),label='Low')
HD_high.plot(figsize=(16, 12),label='High')
plt.xlabel('Date', fontsize=20)
plt.ylabel('Price in USD', fontsize=20)
plt.title("HD High/Low vs Time", fontsize=20)
pylab.legend(loc='upper left',fontsize=20)
```

```python
#The 10-bin equal-width histogram for the Volume attribute. 
hist_gram = HD_data['volume'].hist(bins=10)
```

```python
# A boxplot for the Open and Close/Last attributes.

box = HD_data.boxplot(column=['open','close'])
plt.box()
plt.show()
```

```python
HD_high=HD_data['high']
HD_open=HD_data['open']
HD_high.plot(figsize=(20, 20),label='high')
HD_open.plot(figsize=(20, 20),label='open')
plt.xlabel('Date', fontsize=20)
plt.ylabel('Price in USD', fontsize=20)
plt.title("HD High/open price vs Time", fontsize=20)
pylab.legend(loc='upper left',fontsize=20)
```
